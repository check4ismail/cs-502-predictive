```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 25)

df = pd.read_csv('assignment_3_data.csv')
df
```

```python
# Lowercase all car brands in column x34
df['x34'] = df['x34'].str.lower()
df
```

```python
# In column x68, replace 'sept.' with 'Sept'
df["x68"].replace({'sept.': 'Sept', 'Dev': 'Dec'}, inplace=True)
df
```

```python
# In column x35, make sure days are consistent
df["x35"].replace({
    'monday': 'mon', 
    'sunday': 'sun', 
    'saturday': 'sat', 
    'tuesday': 'tues', 
    'wednesday': 'wed', 
    'thurday': 'thur', 
    'friday': 'fri'}, 
    inplace=True)
df
```

```python
df['x41'] = df['x41'].str.replace('\$', '') # Remove dollar sign from column x41
df['x41'] = df['x41'].str.replace('(', '-') # Replace first paranthesis with negative symbol
df['x41'] = df['x41'].str.replace(')', '') # Remove last paranthesis
df['x41'] = df['x41'].str.replace(',', '') # Remove comma
df['x41'] = df['x41'].astype(float)
df
```

```python
df['x45'] = df['x45'].str.replace('%', '') # Remove percentage symbol
df['x45'] = df['x45'].astype(float) # Convert to float
df
```

```python
df.to_csv('clean_data_assignment_3.csv')
```
